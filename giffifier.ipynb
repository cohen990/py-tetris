```python
lines = open("very-good-game.txt").readlines()
```

```python
lines
```

```python
games = []

marker = '========================================\n'
in_game = False
game = ""
for line in lines:
    if line == marker and in_game:
        game += line
        games.append(game)
        game = ""
        in_game = False
        continue
    if line == marker and not in_game:
        in_game = True
    game += line
        
```
