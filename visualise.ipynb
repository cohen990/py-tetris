```python
import numpy as np
from numpy import array, float32
import matplotlib.pyplot as plot
```

```python
def get_image(node):
    rectified = np.subtract(node, min(node))
    normalised = np.divide(rectified, max(rectified))
    flat_image = np.multiply(normalised, 255)
    image = flat_image.reshape(20, 10)
    return(image)
```

```python
np.set_printoptions(threshold=np.nan)
weights = open("out/32c19cec-ce2a-4562-8b13-8a0794066f79/weights.txt").read()
weights = np.array(eval(weights))
first_layer = weights[0]
first_layer = first_layer.reshape(64, 200)
for node in first_layer:
    image = get_image(node)
    plot.imshow(image, cmap='gray')
    plot.show()
```
