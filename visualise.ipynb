```python
import numpy as np
from numpy import array, float32
import matplotlib.pyplot as plot
```

```python
def normalise(array):
    rectified = np.subtract(array, min(array))
    if max(rectified) == 0:
        return array
    normalised = np.divide(rectified, max(rectified))
    return normalised
```

```python
def get_image(node):
    flat_image = np.multiply(normalise(node), 255)
    image = flat_image.reshape(20, 10)
    return(image)
```

```python
np.set_printoptions(threshold=np.nan)
weights = open("out/3218a454-3b73-4655-aa70-96b0d756ab54/weights.txt").read()
weights = np.array(eval(weights))
```

```python
first_layer = weights[0]
first_layer_reshaped = first_layer.reshape(64, 200)
for node in first_layer_reshaped:
    image = get_image(node)
    plot.imshow(image, cmap='gray')
    plot.show()
```

```python
second_layer = weights[2]
```

```python
images = []
for node in second_layer:
    weighted = np.inner(first_layer, node)
    image = get_image(weighted)
    images.append(image)
```

```python
for image in images:
    plot.imshow(image, cmap='gray')
    plot.show()
```
